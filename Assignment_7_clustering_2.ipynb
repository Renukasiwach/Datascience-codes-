```python
# normalized data function
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

# KMeans

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

KMeans alogithum aims to choose centroids that minimise the inertia,or within_cluster sum of squares criterio.

```python
crime1=pd.read_csv('crime_data.csv')
crime1
```

```python
# normalized data function
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
df_norm=norm_func(crime1.iloc[:,1:])
```

```python
# Elbow curve
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(df_norm)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curve')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()  
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime1['clust']=x
```

```python
crime1
```

```python
crime1.iloc[:,1:5].groupby(crime1.clust).mean()
```

```python
fig, ax=plt.subplots(figsize=(10,6))
crime.groupby(['h_clusterid']).count()['State'].plot(kind='bar')
plt.ylabel('States')
plt.title('KMeans Clustering',fontsize='large',fontweight='bold')
ax.set_xlabel('Clusters',fontsize='large',fontweight='bold')
ax.set_ylabel('States',fontsize='large',fontweight='bold')
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.metrics import silhouette_score
```

```python
EPS=[0.22,0.24,0.26,0.28,0.30]

for n in EPS:
  dbscan=DBSCAN(eps=n ,min_samples=4)
  dbscan.fit(df_norm)
  y=pd.DataFrame(dbscan.fit_predict(df_norm),columns=['clusterid_DBSCAN'])
  print(f'For eps={n}','\n',y['clusterid_DBSCAN'].value_counts())

  #silhouette score
  sil_score=silhouette_score(df_norm, dbscan.labels_)
  print(f'For eps silhouette score={n}','\n',sil_score)
```

When we have value of epsilon=0.28,we are getting 3 clusters silhouette score is more as compared to other dbscan models.

```python
dbscan=DBSCAN(eps=0.28, min_samples=4)
dbscan.fit(df_norm)
```

```python
crime['clusterid_DBSCAN']=dbscan.labels_
crime.head()
```

```python
crime.groupby('clusterid_DBSCAN').agg(['mean']).reset_index()
```

```python
fig, ax=plt.subplots(figsize=(10,6))
crime.groupby(['clusterid_DBSCAN']).count()['State'].plot(kind='bar')
plt.ylabel('States')
plt.title('DBSCAN Clustering',fontsize='large',fontweight='bold')
ax.set_xlabel('Clusters',fontsize='large',fontweight='bold')
ax.set_ylabel('States',fontsize='large',fontweight='bold')
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```
